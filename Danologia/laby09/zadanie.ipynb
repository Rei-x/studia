Stwórz 2 różne macierze, a następnie wykonaj na nich operacje: +, -, *, /, **, //,
%.

```python
macierz_A = np.array([[2, 3], [4, 5], [6, 7]])
macierz_B = np.array([[1, 2], [3, 4], [5, 6]])

dodawanie = macierz_A + macierz_B
odejmowanie = macierz_A - macierz_B
mnozenie = macierz_A * macierz_B
dzielenie = macierz_A / macierz_B
potegowanie = macierz_A ** 2
dzielenie_całkowite = macierz_A // macierz_B
reszta = macierz_A % macierz_B

print("Dodawanie:")
print(dodawanie)
print("\nOdejmowanie:")
print(odejmowanie)
print("\nMnożenie (element-wise):")
print(mnozenie)
print("\nDzielenie:")
print(dzielenie)
print("\nPotęgowanie:")
print(potegowanie)
print("\nDzielenie całkowite:")
print(dzielenie_całkowite)
print("\nReszta z dzielenia:")
print(reszta)
```

Zaimplementuj algorytm sortowania szybkiego, wykorzystując funkcję/metodę
partition().

```python
import random


vector = np.random.randint(-100, 500, 100)

def quicksort(arr: np.ndarray) -> np.ndarray:
    arr_len = len(arr)
    if arr_len <= 1:
        return arr

    pivot = random.randint(0, arr_len - 1)

    partitioned = np.partition(arr, pivot)

    return np.concat((quicksort(partitioned[:pivot]), [partitioned[pivot]],  quicksort(partitioned[pivot+1:]))) 

print("Before")
print(vector)
print("After")
sorted_vector = quicksort(vector)
print(sorted_vector)
print("Is sorted?")
print(np.array_equal(sorted_vector, np.sort(vector)))
```

Napisz funkcję, która wywołując unique() (z odpowiednimi argumentami),
wyznacza modę (dominantę) wartości w podanym wektorze.

```python
vector = np.random.randint(0, 2, 10)

def find_dominant(arr: np.ndarray):
  elements, counts = np.unique(arr, return_counts=True)
  max_count = np.max(counts)

  return elements[np.where(counts == max_count)]

print(vector)
find_dominant(vector)
```

Ramka danych flights zawiera kolumny year i month. Wygneruj na ich
podstawie nową zmienną typu data i czas.

```python
def month_to_num(name: str) -> int:
  month_mapping = {
    "Jan": 1,
    "Feb": 2,
    "Mar": 3,
    "Apr": 4,
    "May": 5,
    "Jun": 6,
    "Jul": 7,
    "Aug": 8,
    "Sep": 9,
    "Oct": 10,
    "Nov": 11,
    "Dec": 12
  }
  return month_mapping.get(name, 0)

new_flights = flights.copy()
new_flights["datetime"] = pd.to_datetime(
  new_flights["year"].astype(str) + "-" + 
  new_flights["month"].apply(month_to_num).astype(str) + "-01"
)

new_flights
```

Podziel ramkę danych irys na dwie rozłączne części: pierwsza 80%
obserwacji, druga 20%.

```python
whole_len = len(iris)
train_len = int(0.8 * whole_len)

eighty_iris = iris[:train_len]
twenty_iris = iris[train_len:]

eighty_iris.shape, twenty_iris.shape
```

Dokonaj standaryzacji wszystkich zmiennych liczbowych w ramce danych iris.

```python
iris_numeric = iris.select_dtypes(include=[np.number])

srednia = np.mean(iris_numeric, axis=0)
odchylenie = np.std(iris_numeric, axis=0)

iris_standardized_numeric = (iris_numeric - srednia) / odchylenie

iris[iris_numeric.columns] = iris_standardized_numeric

np.std(iris_standardized_numeric, axis=0), np.mean(iris_standardized_numeric, axis=0), iris.head()
```

Wyznacz podstawowe statystyki próbkowe (średnia, odchylenie standardowe,
kwartyle) dla liczby pasażerów (flights) w każdym roku z sobna. Przedstaw
wyniki w taki sposób, by zagregowane wartości przechowywane były
w kolumnach, a kolejne lata w wierszach.

```python
flights.groupby('year')['passengers'].agg([
    ('średnia', 'mean'),
    ('odchylenie_standardowe', 'std'),
    ('Q1', lambda x: x.quantile(0.25)),
    ('mediana', 'median'),
    ('Q3', lambda x: x.quantile(0.75))
])
```
