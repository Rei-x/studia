```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

with open('supermarket.csv', 'r') as f:
    transactions = [line.strip().split(',') for line in f]
```

```python
# low, high, department

# One-hot encode the data
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Find frequent itemsets with a minimum support of 20%
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)

# Find the size of each frequent itemset
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

# Find the maximum length among the frequent itemsets
max_length = frequent_itemsets['length'].max()

# Filter for the most numerous frequent itemsets
largest_frequent_itemsets = frequent_itemsets[frequent_itemsets['length'] == max_length]

# Find the itemset with the highest support among the largest ones
most_frequent_largest_itemset = largest_frequent_itemsets.sort_values(by='support', ascending=False).iloc[0]

# Extract the item names and support
item_names = most_frequent_largest_itemset['itemsets']
support = most_frequent_largest_itemset['support']

# Print the results
print("Najliczniejszy zbiór częsty:")
print(f"Produkty: {', '.join(item_names)}")
print(f"Wsparcie: {support:.2f}")
```
