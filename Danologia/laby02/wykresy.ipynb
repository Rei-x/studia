```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Funkcja Rosenbrocka: f(x,y) = (a-x)^2 + b(y-x^2)^2
# Zazwyczaj a=1, b=100
a = 1
b = 100

# Tworzenie siatki punktw
x = np.linspace(-2, 2, 400)
y = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x, y)

# Obliczenie wartoci funkcji Rosenbrocka
Z = (a - X)**2 + b * (Y - X**2)**2

# Tworzenie figury z dwoma wykresami
fig = plt.figure(figsize=(16, 6))

# Wykres 1: Powierzchnia 3D
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.8, 
                        edgecolor='none', vmin=0, vmax=2000)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f(x, y)')
ax1.set_title('Funkcja Rosenbrocka - widok 3D')
ax1.view_init(elev=30, azim=45)
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

# Wykres 2: Mapa konturowa (widok z gry)
ax2 = fig.add_subplot(122)
# Uywamy skali logarytmicznej dla lepszej wizualizacji
levels = np.logspace(0, 3.5, 35)
contour = ax2.contour(X, Y, Z, levels=levels, cmap='coolwarm', linewidths=1)
contourf = ax2.contourf(X, Y, Z, levels=levels, cmap='coolwarm', alpha=0.6)
ax2.plot(1, 1, 'r*', markersize=20, label='Minimum globalne (1, 1)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Funkcja Rosenbrocka - mapa konturowa')
ax2.legend()
ax2.grid(True, alpha=0.3)
fig.colorbar(contourf, ax=ax2, shrink=0.8)

plt.tight_layout()
plt.show()

print(f"Funkcja Rosenbrocka: f(x,y) = (a-x) + b(y-x)")
print(f"Parametry: a = {a}, b = {b}")
print(f"Minimum globalne: f(1, 1) = 0")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 300)
y = np.linspace(-10, 10, 300)
X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)

Z = (np.cos(R)**2) * np.exp(-0.3 * R)

fig = plt.figure(figsize=(16, 6))

ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='plasma', alpha=0.9, 
                        edgecolor='none', antialiased=True)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('Funkcja z = cos(R) * exp(-0.3R) - widok 3D')
ax1.view_init(elev=25, azim=45)
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

ax2 = fig.add_subplot(122)
levels = 20
contourf = ax2.contourf(X, Y, Z, levels=levels, cmap='plasma', alpha=0.8)
contour = ax2.contour(X, Y, Z, levels=levels, colors='black', linewidths=0.5, alpha=0.3)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Funkcja z = cos(R) * exp(-0.3R) - mapa konturowa')
ax2.set_aspect('equal')
ax2.grid(True, alpha=0.3, linestyle=':')
fig.colorbar(contourf, ax=ax2, shrink=0.8)

plt.tight_layout()
plt.show()

print("Funkcja: z = cos(R) * exp(-0.3*R)")
print("gdzie R = (X + Y)")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

# Parametry czasowe
t = np.linspace(0, 2, 2000)  # 2 sekundy, 2000 prbek
fs = 1000  # czstotliwo prbkowania (Hz)

# 1. Fala prostoktna (square wave)
square_wave = signal.square(2 * np.pi * 5 * t)  # czstotliwo 5 Hz

# 2. Przebieg pioksztatny (sawtooth wave)
sawtooth_wave = signal.sawtooth(2 * np.pi * 5 * t)  # czstotliwo 5 Hz

# 3. Przebieg wiergotliwy (chirp) - czstotliwo narastajca
chirp_wave = signal.chirp(t, f0=5, f1=50, t1=2, method='linear')

# Obliczenie periodogramw (Widmo Gstoci Mocy)
# Periodogram dla fali prostoktnej
f_square, Pxx_square = signal.periodogram(square_wave, fs)

# Periodogram dla przebiegu pioksztatnego
f_sawtooth, Pxx_sawtooth = signal.periodogram(sawtooth_wave, fs)

# Periodogram dla przebiegu wiergotliwego
f_chirp, Pxx_chirp = signal.periodogram(chirp_wave, fs)

# Tworzenie figury z 3 wierszami i 2 kolumnami
fig, axes = plt.subplots(3, 2, figsize=(16, 12))

# Wiersz 1: Fala prostoktna
# Przebieg czasowy
axes[0, 0].plot(t, square_wave, color='blue', linewidth=1.5)
axes[0, 0].set_title('Fala prostoktna - Przebieg czasowy', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Czas [s]')
axes[0, 0].set_ylabel('Amplituda')
axes[0, 0].grid(True, alpha=0.3, linestyle='--')
axes[0, 0].set_ylim(-1.5, 1.5)
axes[0, 0].axhline(0, color='black', linewidth=0.5)

# Widmo mocy (WGM)
axes[0, 1].semilogy(f_square, Pxx_square, color='blue', linewidth=1.5)
axes[0, 1].set_title('Fala prostoktna - Widmo Gstoci Mocy (WGM)', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Czstotliwo [Hz]')
axes[0, 1].set_ylabel('Amplituda')
axes[0, 1].grid(True, alpha=0.3, linestyle='--')
axes[0, 1].set_xlim(0, 100)

# Wiersz 2: Przebieg pioksztatny
# Przebieg czasowy
axes[1, 0].plot(t, sawtooth_wave, color='green', linewidth=1.5)
axes[1, 0].set_title('Przebieg pioksztatny - Przebieg czasowy', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Czas [s]')
axes[1, 0].set_ylabel('Amplituda')
axes[1, 0].grid(True, alpha=0.3, linestyle='--')
axes[1, 0].set_ylim(-1.5, 1.5)
axes[1, 0].axhline(0, color='black', linewidth=0.5)

# Widmo mocy (WGM)
axes[1, 1].semilogy(f_sawtooth, Pxx_sawtooth, color='green', linewidth=1.5)
axes[1, 1].set_title('Przebieg pioksztatny - Widmo Gstoci Mocy (WGM)', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Czstotliwo [Hz]')
axes[1, 1].set_ylabel('Amplituda')
axes[1, 1].grid(True, alpha=0.3, linestyle='--')
axes[1, 1].set_xlim(0, 100)

# Wiersz 3: Przebieg wiergotliwy (chirp)
# Przebieg czasowy
axes[2, 0].plot(t, chirp_wave, color='red', linewidth=1)
axes[2, 0].set_title('Przebieg wiergotliwy (Chirp) - Przebieg czasowy', fontsize=12, fontweight='bold')
axes[2, 0].set_xlabel('Czas [s]')
axes[2, 0].set_ylabel('Amplituda')
axes[2, 0].grid(True, alpha=0.3, linestyle='--')
axes[2, 0].set_ylim(-1.5, 1.5)
axes[2, 0].axhline(0, color='black', linewidth=0.5)

# Widmo mocy (WGM)
axes[2, 1].semilogy(f_chirp, Pxx_chirp, color='red', linewidth=1.5)
axes[2, 1].set_title('Przebieg wiergotliwy (Chirp) - Widmo Gstoci Mocy (WGM)', fontsize=12, fontweight='bold')
axes[2, 1].set_xlabel('Czstotliwo [Hz]')
axes[2, 1].set_ylabel('Amplituda')
axes[2, 1].grid(True, alpha=0.3, linestyle='--')
axes[2, 1].set_xlim(0, 100)

plt.tight_layout()
plt.show()

print("Analiza widmowa sygnaw:")
print("\n1. Fala prostoktna:")
print("   - Widmo zawiera czstotliwo podstawow (5 Hz) i nieparzyste harmoniczne")
print("   - Charakterystyczne dla sygnau prostoktnego")
print("\n2. Przebieg pioksztatny:")
print("   - Widmo zawiera wszystkie harmoniczne (parzyste i nieparzyste)")
print("   - Amplitudy malej zgodnie z 1/n")
print("\n3. Przebieg wiergotliwy (Chirp):")
print("   - Energia rozoona w zakresie 5-50 Hz")
print("   - Szerokopasmowy charakter sygnau o zmiennej czstotliwoci")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import electrocardiogram
from scipy.signal import find_peaks

# Wczytanie danych EKG
ecg = electrocardiogram()

# Czstotliwo prbkowania dla danych EKG to 360 Hz
fs = 360  # Hz
time = np.arange(len(ecg)) / fs

# Wykrywanie szczytw (peaks)
# Parametry: height - minimalna wysoko szczytu, distance - minimalna odlego midzy szczytami
peaks, properties = find_peaks(ecg, height=0.5, distance=150)

# Tworzenie figury z dwoma wykresami
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# Wykres 1: Peny sygna EKG
axes[0].plot(time, ecg, color='blue', linewidth=0.8, label='Sygna EKG')
axes[0].plot(time[peaks], ecg[peaks], 'ro', markersize=6, label=f'Wykryte szczyty ({len(peaks)})')
axes[0].set_title('Elektrokardiogram (EKG) - peny sygna', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Czas [s]', fontsize=12)
axes[0].set_ylabel('Amplituda [mV]', fontsize=12)
axes[0].grid(True, alpha=0.3, linestyle='--')
axes[0].legend(loc='upper right', fontsize=10)
axes[0].axhline(0, color='black', linewidth=0.5, alpha=0.5)

# Wykres 2: Fragment sygnau EKG (pierwsze 5 sekund) dla lepszej widocznoci
time_limit = 5  # sekundy
indices = time <= time_limit

axes[1].plot(time[indices], ecg[indices], color='blue', linewidth=1.5, label='Sygna EKG')
peaks_in_range = peaks[peaks < np.sum(indices)]
axes[1].plot(time[peaks_in_range], ecg[peaks_in_range], 'ro', markersize=10, 
            label=f'Wykryte szczyty', zorder=5)
axes[1].set_title('Elektrokardiogram (EKG) - fragment (0-5s)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Czas [s]', fontsize=12)
axes[1].set_ylabel('Amplituda [mV]', fontsize=12)
axes[1].grid(True, alpha=0.3, linestyle='--')
axes[1].legend(loc='upper right', fontsize=10)
axes[1].axhline(0, color='black', linewidth=0.5, alpha=0.5)

plt.tight_layout()
plt.show()

# Obliczenie statystyk
peak_times = time[peaks]
if len(peak_times) > 1:
    rr_intervals = np.diff(peak_times)  # R-R intervals (odstpy midzy szczytami)
    heart_rate = 60.0 / rr_intervals  # Ttno w uderzeniach na minut
    
    print("Statystyki sygnau EKG:")
    print(f"Dugo sygnau: {time[-1]:.2f} sekund")
    print(f"Liczba wykrytych szczytw (zaamkw R): {len(peaks)}")
    print(f"\nStatystyki ttna:")
    print(f"  rednie ttno: {np.mean(heart_rate):.1f} BPM (uderze/minut)")
    print(f"  Minimalne ttno: {np.min(heart_rate):.1f} BPM")
    print(f"  Maksymalne ttno: {np.max(heart_rate):.1f} BPM")
    print(f"  Odchylenie standardowe: {np.std(heart_rate):.1f} BPM")
    print(f"\nredni interwa R-R: {np.mean(rr_intervals)*1000:.1f} ms")
else:
    print(f"Wykryto {len(peaks)} szczytw w sygnale EKG")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

# Parametry czasowe
t = np.linspace(0, 2, 2000)  # 2 sekundy, 2000 prbek
fs = 1000  # czstotliwo prbkowania (Hz)

# 1. Fala prostoktna (lewy grny)
square_wave = signal.square(2 * np.pi * 5 * t)

# 2. Przebieg pioksztatny (lewy dolny)
sawtooth_wave = signal.sawtooth(2 * np.pi * 5 * t)

# 3. Histogram - dane losowe (prawy grny)
np.random.seed(42)
hist_data = np.random.randn(1000)

# 4. Superpozycja 3 sinusw (prawy dolny)
# Czstotliwoci: 10 Hz, 25 Hz, 40 Hz
sin1 = np.sin(2 * np.pi * 10 * t)
sin2 = 0.7 * np.sin(2 * np.pi * 25 * t)
sin3 = 0.5 * np.sin(2 * np.pi * 40 * t)
superposition = sin1 + sin2 + sin3

fig1, axes1 = plt.subplots(2, 2, figsize=(16, 10))
fig1.suptitle('Sygnay czasowe', fontsize=16, fontweight='bold', y=1.00)

# Lewy grny: Fala prostoktna
axes1[0, 0].plot(t, square_wave, color='blue', linewidth=1.5)
axes1[0, 0].set_title('Fala prostoktna', fontsize=13, fontweight='bold')
axes1[0, 0].set_xlabel('Czas [s]')
axes1[0, 0].set_ylabel('Amplituda')
axes1[0, 0].grid(True, alpha=0.3, linestyle='--')
axes1[0, 0].set_ylim(-1.5, 1.5)
axes1[0, 0].axhline(0, color='black', linewidth=0.5)

# Lewy dolny: Przebieg pioksztatny
axes1[1, 0].plot(t, sawtooth_wave, color='green', linewidth=1.5)
axes1[1, 0].set_title('Przebieg pioksztatny', fontsize=13, fontweight='bold')
axes1[1, 0].set_xlabel('Czas [s]')
axes1[1, 0].set_ylabel('Amplituda')
axes1[1, 0].grid(True, alpha=0.3, linestyle='--')
axes1[1, 0].set_ylim(-1.5, 1.5)
axes1[1, 0].axhline(0, color='black', linewidth=0.5)

# Prawy grny: Histogram
axes1[0, 1].hist(hist_data, bins=30, color='purple', alpha=0.7, edgecolor='black')
axes1[0, 1].set_title('Histogram danych losowych', fontsize=13, fontweight='bold')
axes1[0, 1].set_xlabel('Warto')
axes1[0, 1].set_ylabel('Czsto')
axes1[0, 1].grid(True, alpha=0.3, linestyle='--')

# Prawy dolny: Superpozycja 3 sinusw
axes1[1, 1].plot(t, superposition, color='red', linewidth=1)
axes1[1, 1].set_title('Superpozycja 3 sinusw (10 Hz, 25 Hz, 40 Hz)', fontsize=13, fontweight='bold')
axes1[1, 1].set_xlabel('Czas [s]')
axes1[1, 1].set_ylabel('Amplituda')
axes1[1, 1].grid(True, alpha=0.3, linestyle='--')
axes1[1, 1].axhline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()


# Metoda Welcha dzieli sygna na segmenty i urednia ich periodogramy
f_square, Pxx_square = signal.welch(square_wave, fs, nperseg=256)
f_sawtooth, Pxx_sawtooth = signal.welch(sawtooth_wave, fs, nperseg=256)
f_hist, Pxx_hist = signal.welch(hist_data, fs, nperseg=256)
f_super, Pxx_super = signal.welch(superposition, fs, nperseg=256)

fig2, axes2 = plt.subplots(2, 2, figsize=(16, 10))
fig2.suptitle('Widma Gstoci Mocy (WGM) - Metoda Welcha', fontsize=16, fontweight='bold', y=1.00)

# Lewy grny: WGM fali prostoktnej
axes2[0, 0].semilogy(f_square, Pxx_square, color='blue', linewidth=1.5)
axes2[0, 0].set_title('WGM - Fala prostoktna', fontsize=13, fontweight='bold')
axes2[0, 0].set_xlabel('Czstotliwo [Hz]')
axes2[0, 0].set_ylabel('Amplituda')
axes2[0, 0].grid(True, alpha=0.3, linestyle='--')
axes2[0, 0].set_xlim(0, 100)

# Lewy dolny: WGM przebiegu pioksztatnego
axes2[1, 0].semilogy(f_sawtooth, Pxx_sawtooth, color='green', linewidth=1.5)
axes2[1, 0].set_title('WGM - Przebieg pioksztatny', fontsize=13, fontweight='bold')
axes2[1, 0].set_xlabel('Czstotliwo [Hz]')
axes2[1, 0].set_ylabel('Amplituda')
axes2[1, 0].grid(True, alpha=0.3, linestyle='--')
axes2[1, 0].set_xlim(0, 100)

# Prawy grny: WGM danych losowych (szum biay)
axes2[0, 1].semilogy(f_hist, Pxx_hist, color='purple', linewidth=1.5)
axes2[0, 1].set_title('WGM - Dane losowe (szum)', fontsize=13, fontweight='bold')
axes2[0, 1].set_xlabel('Czstotliwo [Hz]')
axes2[0, 1].set_ylabel('Amplituda')
axes2[0, 1].grid(True, alpha=0.3, linestyle='--')
axes2[0, 1].set_xlim(0, 100)

# Prawy dolny: WGM superpozycji 3 sinusw
axes2[1, 1].semilogy(f_super, Pxx_super, color='red', linewidth=1.5)
axes2[1, 1].set_title('WGM - Superpozycja 3 sinusw', fontsize=13, fontweight='bold')
axes2[1, 1].set_xlabel('Czstotliwo [Hz]')
axes2[1, 1].set_ylabel('Amplituda')
axes2[1, 1].grid(True, alpha=0.3, linestyle='--')
axes2[1, 1].set_xlim(0, 100)

# Dodanie pionowych linii dla czstotliwoci sinusw
axes2[1, 1].axvline(10, color='orange', linestyle=':', linewidth=2, alpha=0.7, label='10 Hz')
axes2[1, 1].axvline(25, color='cyan', linestyle=':', linewidth=2, alpha=0.7, label='25 Hz')
axes2[1, 1].axvline(40, color='magenta', linestyle=':', linewidth=2, alpha=0.7, label='40 Hz')
axes2[1, 1].legend(loc='upper right')

plt.tight_layout()
plt.show()
```
