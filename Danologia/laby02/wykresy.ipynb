```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 3.6 * np.pi, 400)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='purple', linewidth=2)

plt.title('Wykres funkcji sinus')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 500)

y1 = np.sin(x) * np.exp(-(x/5)**2)
y2 = np.cos(x) * np.exp(-(x/2)**2)

plt.figure(figsize=(12, 7))

plt.plot(x, y1, label='y = sin(x) * exp(-(x/5))', color='blue', linewidth=2)
plt.plot(x, y2, label='y2 = cos(x) * exp(-(x/2))', color='green', linewidth=2)

plt.title('Wykresy funkcji')
plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dane dla wykresu liniowego z punktami (grny lewy rg)
x_line = np.array([0, 1.5, 2.5, 4, 5, 6.5, 7.5, 9])
y_line = np.array([1, 4, 2, 5.5, 3.5, 6, 5, 6.5])

# Dane dla wykresu sinus (dolny lewy rg)
x_sin = np.linspace(0, 3.6 * np.pi, 400)
y_sin = np.sin(x_sin)

# Dane dla dwch funkcji tumionych (bd uyte w legendzie w dolnym lewym rogu)
x_damped = np.linspace(-10, 10, 500)
y_damped_1 = np.sin(x_damped) * np.exp(-(x_damped/5) ** 2)
y_damped_2 = np.cos(x_damped) * np.exp(-(x_damped/2) ** 2)

# Dane do histogramu (grny prawy rg)
hist_data = np.random.randn(100)

# Dane do wykresu punktowego (dolny prawy rg)
scatter_x = np.random.uniform(-5, 5, 20)
scatter_y = np.random.uniform(-5, 5, 20)

# Tworzenie siatki 2x2
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Pierwszy wykres: linia z punktami (0,0)
ax00 = axes[0, 0]
ax00.plot(x_line, y_line, color='black', linewidth=2)
ax00.scatter(x_line, y_line, color='red', s=50)
ax00.set_title('Wykres liniowy z punktami')
ax00.set_xlabel('X')
ax00.set_ylabel('Y')
ax00.grid(True, linestyle=':', alpha=0.3)

# Drugi wykres: funkcje tumione (1,0)
ax10 = axes[1, 0]
ax10.plot(x_damped, y_damped_1, label='sin(x) * exp(-(x/5))', color='blue', linewidth=2)
ax10.plot(x_damped, y_damped_2, label='cos(x) * exp(-(x/2))', color='green', linewidth=2)
ax10.set_title('Funkcje tumione')
ax10.set_xlabel('X')
ax10.set_ylabel('Warto funkcji')
ax10.grid(True, linestyle='--', alpha=0.6)
ax10.legend()

# Trzeci wykres: histogram (0,1)
ax01 = axes[0, 1]
ax01.hist(hist_data, bins=15, color='purple', alpha=0.7, edgecolor='black')
ax01.set_title('Histogram 100 losowych wartoci')
ax01.set_xlabel('Wartoci')
ax01.set_ylabel('Czsto')
ax01.grid(True, linestyle='--', alpha=0.4)

# Czwarty wykres: losowe punkty (1,1)
ax11 = axes[1, 1]
ax11.scatter(scatter_x, scatter_y, color='orange', s=60, edgecolor='black', marker="*")
ax11.set_title('Losowe 20 punktw')
ax11.set_xlabel('X')
ax11.set_ylabel('Y')
ax11.grid(True, linestyle=':', alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = X**2 + X * Y**2 + Y**3

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8, edgecolor='none')

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Wykres 3D funkcji: z = x + xy + y')

ax.view_init(elev=25, azim=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Funkcja Rosenbrocka: f(x,y) = (a-x)^2 + b(y-x^2)^2
# Zazwyczaj a=1, b=100
a = 1
b = 100

# Tworzenie siatki punktw
x = np.linspace(-2, 2, 400)
y = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x, y)

# Obliczenie wartoci funkcji Rosenbrocka
Z = (a - X)**2 + b * (Y - X**2)**2

# Tworzenie figury z dwoma wykresami
fig = plt.figure(figsize=(16, 6))

# Wykres 1: Powierzchnia 3D
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='coolwarm', alpha=0.8, 
                        edgecolor='none', vmin=0, vmax=2000)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('f(x, y)')
ax1.set_title('Funkcja Rosenbrocka - widok 3D')
ax1.view_init(elev=30, azim=45)
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

# Wykres 2: Mapa konturowa (widok z gry)
ax2 = fig.add_subplot(122)
# Uywamy skali logarytmicznej dla lepszej wizualizacji
levels = np.logspace(0, 3.5, 35)
contour = ax2.contour(X, Y, Z, levels=levels, cmap='coolwarm', linewidths=1)
contourf = ax2.contourf(X, Y, Z, levels=levels, cmap='coolwarm', alpha=0.6)
ax2.plot(1, 1, 'r*', markersize=20, label='Minimum globalne (1, 1)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Funkcja Rosenbrocka - mapa konturowa')
ax2.legend()
ax2.grid(True, alpha=0.3)
fig.colorbar(contourf, ax=ax2, shrink=0.8)

plt.tight_layout()
plt.show()

print(f"Funkcja Rosenbrocka: f(x,y) = (a-x) + b(y-x)")
print(f"Parametry: a = {a}, b = {b}")
print(f"Minimum globalne: f(1, 1) = 0")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 300)
y = np.linspace(-10, 10, 300)
X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)

Z = (np.cos(R)**2) * np.exp(-0.3 * R)

fig = plt.figure(figsize=(16, 6))

ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='plasma', alpha=0.9, 
                        edgecolor='none', antialiased=True)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('Funkcja z = cos(R) * exp(-0.3R) - widok 3D')
ax1.view_init(elev=25, azim=45)
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

ax2 = fig.add_subplot(122)
levels = 20
contourf = ax2.contourf(X, Y, Z, levels=levels, cmap='plasma', alpha=0.8)
contour = ax2.contour(X, Y, Z, levels=levels, colors='black', linewidths=0.5, alpha=0.3)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Funkcja z = cos(R) * exp(-0.3R) - mapa konturowa')
ax2.set_aspect('equal')
ax2.grid(True, alpha=0.3, linestyle=':')
fig.colorbar(contourf, ax=ax2, shrink=0.8)

plt.tight_layout()
plt.show()

print("Funkcja: z = cos(R) * exp(-0.3*R)")
print("gdzie R = (X + Y)")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(42)

n_points = 200

cluster1_x = np.random.normal(2, 0.8, n_points)
cluster1_y = np.random.normal(2, 0.8, n_points)
cluster1_z = np.random.normal(2, 0.8, n_points)

cluster2_x = np.random.normal(-3, 0.9, n_points)
cluster2_y = np.random.normal(1, 0.9, n_points)
cluster2_z = np.random.normal(4, 0.9, n_points)

cluster3_x = np.random.normal(1, 0.7, n_points)
cluster3_y = np.random.normal(-2, 0.7, n_points)
cluster3_z = np.random.normal(-1, 0.7, n_points)

fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')

scatter1 = ax.scatter(cluster1_x, cluster1_y, cluster1_z, 
                     c='red', marker='o', s=30, alpha=0.6, 
                     label='Klaster 1', edgecolors='darkred', linewidth=0.5)

scatter2 = ax.scatter(cluster2_x, cluster2_y, cluster2_z, 
                     c='green', marker='^', s=30, alpha=0.6, 
                     label='Klaster 2', edgecolors='darkgreen', linewidth=0.5)

scatter3 = ax.scatter(cluster3_x, cluster3_y, cluster3_z, 
                     c='blue', marker='s', s=30, alpha=0.6, 
                     label='Klaster 3', edgecolors='darkblue', linewidth=0.5)

ax.scatter([2], [2], [2], c='darkred', marker='*', s=300, 
          edgecolors='black', linewidth=2, label='Centroidy')
ax.scatter([-3], [1], [4], c='darkgreen', marker='*', s=300, 
          edgecolors='black', linewidth=2)
ax.scatter([1], [-2], [-1], c='darkblue', marker='*', s=300, 
          edgecolors='black', linewidth=2)

ax.set_xlabel('X', fontsize=12, fontweight='bold')
ax.set_ylabel('Y', fontsize=12, fontweight='bold')
ax.set_zlabel('Z', fontsize=12, fontweight='bold')
ax.set_title('Trzy klastry danych w przestrzeni 3D', fontsize=14, fontweight='bold')

ax.legend(loc='upper left', fontsize=10)

ax.view_init(elev=20, azim=45)

ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("Statystyki klastrw:")
print(f"Klaster 1 (czerwony): {n_points} punktw, centrum w (2, 2, 2)")
print(f"Klaster 2 (zielony): {n_points} punktw, centrum w (-3, 1, 4)")
print(f"Klaster 3 (niebieski): {n_points} punktw, centrum w (1, -2, -1)")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

# Parametry czasowe
t = np.linspace(0, 2, 2000)  # 2 sekundy, 2000 prbek

# 1. Fala prostoktna (square wave)
square_wave = signal.square(2 * np.pi * 5 * t)  # czstotliwo 5 Hz

# 2. Przebieg pioksztatny (sawtooth wave)
sawtooth_wave = signal.sawtooth(2 * np.pi * 5 * t)  # czstotliwo 5 Hz

# 3. Przebieg wiergotliwy (chirp) - czstotliwo narastajca
# Chirp od 5 Hz do 50 Hz w cigu 2 sekund
chirp_wave = signal.chirp(t, f0=5, f1=50, t1=2, method='linear')

# Tworzenie figury z trzema wykresami
fig, axes = plt.subplots(3, 1, figsize=(14, 10))

# Wykres 1: Fala prostoktna
axes[0].plot(t, square_wave, color='blue', linewidth=1.5)
axes[0].set_title('Fala prostoktna (Square Wave)', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Czas [s]')
axes[0].set_ylabel('Amplituda')
axes[0].grid(True, alpha=0.3, linestyle='--')
axes[0].set_ylim(-1.5, 1.5)
axes[0].axhline(0, color='black', linewidth=0.5)

# Wykres 2: Przebieg pioksztatny
axes[1].plot(t, sawtooth_wave, color='green', linewidth=1.5)
axes[1].set_title('Przebieg pioksztatny (Sawtooth Wave)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Czas [s]')
axes[1].set_ylabel('Amplituda')
axes[1].grid(True, alpha=0.3, linestyle='--')
axes[1].set_ylim(-1.5, 1.5)
axes[1].axhline(0, color='black', linewidth=0.5)

# Wykres 3: Przebieg wiergotliwy (chirp)
axes[2].plot(t, chirp_wave, color='red', linewidth=1)
axes[2].set_title('Przebieg wiergotliwy - Chirp (5 Hz  50 Hz)', fontsize=13, fontweight='bold')
axes[2].set_xlabel('Czas [s]')
axes[2].set_ylabel('Amplituda')
axes[2].grid(True, alpha=0.3, linestyle='--')
axes[2].set_ylim(-1.5, 1.5)
axes[2].axhline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()

print("Wygenerowane sygnay z scipy.signal:")
print("1. Fala prostoktna - sygna przeczajcy si midzy +1 a -1")
print("2. Przebieg pioksztatny - liniowo narastajcy sygna")
print("3. Przebieg wiergotliwy (chirp) - sygna o narastajcej czstotliwoci")
```
