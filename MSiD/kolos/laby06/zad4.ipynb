```python
import numpy as np
from scipy.optimize import minimize

# Definicja funkcji celu
def objective(x):
    return x[0]**2 + x[1]**2

# Definicja ograniczenia
def constraint(x):
    return x[0] + x[1] - 2

# Funkcja kary
def penalty_function(x, lambd):
    return objective(x) + 0.5 * lambd * (constraint(x))**2

# Początkowe wartości zmiennych
x0 = [0, 4]

# Początkowy współczynnik kary
lambd = 1
# Współczynnik zwiększenia kary
penalty_increase = 10
# Maksymalna liczba iteracji
max_iterations = 10

# Pętla iteracyjna
for i in range(max_iterations):
    # Minimalizacja funkcji kary
    result = minimize(penalty_function, x0, args=(lambd), method='SLSQP')
    
    # Aktualizacja wartości zmiennych
    x0 = result.x
    
    # Zwiększenie współczynnika kary
    lambd *= penalty_increase
    
    # Wyświetlenie wyników dla każdej iteracji
    print(f"Iteracja {i+1}:")
    print("Współczynnik kary:", lambd)
    print("Optymalne wartości zmiennych: x1 =", result.x[0], ", x2 =", result.x[1])
    print("Wartość funkcji celu w minimum:", objective(result.x))
    print("Wartość funkcji ograniczenia:", constraint(result.x))
    print("-" * 50)

# Wyniki końcowe
print("Końcowe optymalne wartości zmiennych: x1 =", result.x[0], ", x2 =", result.x[1])
print("Końcowa wartość funkcji celu w minimum:", objective(result.x))
```

```python
import numpy as np
from scipy.optimize import minimize

# Definicja funkcji celu
def objective(x):
    return (x[0] - 1) * (x[1] - 1) * x[0] * x[1]

# Definicja ograniczenia
def constraint(x):
    return (x[0] - 1)**2 + x[1]**2

# Funkcja kary
def penalty_function(x, lambd):
    return objective(x) + lambd * max(0, constraint(x) - 1)**2

# Początkowe wartości zmiennych
x0 = [-1.6, -0.1]

# Początkowy współczynnik kary
lambd = 0.5
# Współczynnik zwiększenia kary
penalty_increase = 2
# Maksymalna liczba iteracji
max_iterations = 10
# Espilon


# Pętla iteracyjna
for i in range(max_iterations):
    # Minimalizacja funkcji kary
    result = minimize(penalty_function, x0, args=(lambd), method='COBYLA')
    
    # Aktualizacja wartości zmiennych
    x0 = result.x
    
    # Zwiększenie współczynnika kary
    lambd *= penalty_increase
    
    # Wyświetlenie wyników dla każdej iteracji
    print(f"Iteracja {i+1}:")
    print("Współczynnik kary:", lambd)
    print("Optymalne wartości zmiennych: x1 =", result.x[0], ", x2 =", result.x[1])
    print("Wartość funkcji celu w minimum:", objective(result.x))
    print("Wartość funkcji ograniczenia:", constraint(result.x) - 1)
    print("-" * 50)

# Wyniki końcowe
print("Końcowe optymalne wartości zmiennych: x1 =", result.x[0], ", x2 =", result.x[1])
print("Końcowa wartość funkcji celu w minimum:", objective(result.x))
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from typing import Union

# Definicja zmiennej i funkcji celu
x = sp.symbols('x')
f = (x + 1) * x * (x - 1)

# Definicja ograniczeń
g1 = -x - 1
g2 = x - 1

# Parametry metody
alpha: float = 0.01  # stały krok
lambd: float = 1  # współczynnik kary
max_iter: int = 100  # maksymalna liczba iteracji
stop_condition: float = 1e-2  # tolerancja zbieżności

# Gradient funkcji celu
grad_f = sp.diff(f, x)

# Funkcja kary zewnętrznej
def penalty_function(x_val: Union[float, sp.Basic], lambd: float) -> float:
    penalty = lambd * (max(0, g1.subs(x, x_val))**2 + max(0, g2.subs(x, x_val))**2)
    return penalty

# Funkcja celu z karą
def objective_function(x_val: Union[float, sp.Basic], beta: float) -> float:
    return float(f.subs(x, x_val) + penalty_function(x_val, beta))

# Gradient funkcji celu z karą
def gradient(x_val: Union[float, sp.Basic], beta: float) -> float:
    grad = grad_f.subs(x, x_val)
    if g1.subs(x, x_val) > 0:
        grad += beta * 2 * (g1.subs(x, x_val)) * sp.diff(g1, x).subs(x, x_val)
    if g2.subs(x, x_val) > 0:
        grad += beta * 2 * (g2.subs(x, x_val)) * sp.diff(g2, x).subs(x, x_val)
    return float(grad)

# Metoda największego spadku
def steepest_descent(x0: float, beta: float, alpha: float, max_iter: int, stop_condition: float) -> float:
    x_val = x0
    x_vals = [x_val]  # Lista do przechowywania wartości x w każdej iteracji
    for i in range(max_iter):
        grad = gradient(x_val, beta)
        if np.linalg.norm(grad) <= stop_condition:
            break
        x_val = x_val - alpha * grad
        x_vals.append(x_val)
        print(f"Iteracja {i+1}: x = {x_val}, f(x) = {objective_function(x_val, beta)}")
        print(f"Wartość funkcji ograniczenia g1: {g1.subs(x, x_val)}, g2: {g2.subs(x, x_val)}")
    return x_val, x_vals

# Porównanie wyników dla różnych punktów startowych
x_start_points = [-0.5]
for x0 in x_start_points:
    x_min, x_vals = steepest_descent(x0, lambd, alpha, max_iter, stop_condition)
    print(f"Punkt startowy: {x0}, minimum: {x_min}, wartość funkcji: {objective_function(x_min, lambd)}")
    print(f"Wartość funkcji ograniczenia g1: {g1.subs(x, x_min)}, g2: {g2.subs(x, x_min)}")

    # Wykres
    plt.plot(range(len(x_vals)), x_vals, marker='o', label=f'Punkt startowy: {x0}')

plt.xlabel('Iteracja')
plt.ylabel('Wartość x')
plt.title('Przemieszczanie się punktu po każdej iteracji')
plt.legend()
plt.grid(True)
plt.show()
```
